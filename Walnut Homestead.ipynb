```python
import pandas as pd
import folium
import folium.plugins as plugins
```

```python
walnut_2019 = pd.read_csv('2019_walnut.csv')
```

```python
walnut_2019.columns
```

```python
homesteads = walnut_2019[walnut_2019['Homestead Flag'] == 'Y'].reset_index()
print(len(homesteads))
homesteads.head(10)
```

```python
homesteads.AUD_latitude.isnull().sum()
```

```python
print(homesteads[homesteads.AUD_latitude.isnull() == True])
```

```python
homesteads.AUD_latitude = homesteads.AUD_latitude.fillna(homesteads.latitude)
homesteads.AUD_longitude = homesteads.AUD_longitude.fillna(homesteads.longitude)
homesteads.AUD_latitude.isnull().sum()
```

```python
map1 = folium.Map(
    location=[39.1255405,-84.476745],
    tiles='cartodbpositron',
    zoom_start=15,
)

homesteads.apply(lambda row:folium.Marker(location=[row["AUD_latitude"], row["AUD_longitude"]], popup=row['Owner Name1'] + '\n' + str(row['Location House Number']) + ' ' + row['Location Street Name']).add_to(map1), axis=1)
map1
```
