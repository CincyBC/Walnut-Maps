```python
import pandas as pd
import folium
import folium.plugins as plugins
```

```python
walnut_2019 = pd.read_csv('2019_walnut.csv', index_col='Parcel Number')
```

```python
walnut_2019 = walnut_2019.dropna(axis=0, subset=['latitude', 'longitude'])
residential_vacants = walnut_2019[walnut_2019['Property Class Code'] == 500.]
```

Residential Vacants are plots of land that are zoned for residential use and don't have anything built on them. The below map has 2019 tax data (minus plots that could not be geocoded) for the properties listed as Walnut Hills and East Walnut Hills from the Auditor's database. Clicking on a marker will show you the name of the owner and the house number/street name for each property.
N.B. The data is still being cleaned, so incorrectly geocoded properties still exist. They will disappear as the dataset is cleaned.

```python
map1 = folium.Map(
    location=[39.1255405,-84.476745],
    tiles='cartodbpositron',
    zoom_start=15,
)

residential_vacants.apply(lambda row:folium.Marker(location=[row["latitude"], row["longitude"]], popup=row['Owner Name1'] + '\n' + str(row['Location House Number']) + ' ' + row['Location Street Name']).add_to(map1), axis=1)
map1
```

```python
walnut_2019 = walnut_2019.dropna(axis=0, subset=['latitude', 'longitude'])
single_family = walnut_2019[walnut_2019['Property Class Code'] == 510]
```

The below map is all of the single family properties in the area. This one does not have any information attached to the markers.
N.B. Same note as above.

```python
map1 = folium.Map(
    location=[39.1255405,-84.476745],
    tiles='cartodbpositron',
    zoom_start=15,
)

single_family.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```
